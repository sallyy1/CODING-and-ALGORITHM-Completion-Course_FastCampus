## 과제3.

Generator ABC를 구현하여 아래와 같은 동작을 하는 제네레이터 클래스를 작성하시오.

- 클래스의 이름은 Factorization으로 한다.
- 클래스 생성 시 정수 n을 입력받는다.
- 제네레이터는 n번 반복 가능하며, i번째 출력은 아래와 같다.
  - 숫자 i는 1부터 시작하며, 마지막 숫자는 n이다.
  - 숫자 i를 소인수분해한 결과를 딕셔너리로 출력한다. 딕셔너리의 key는 소인수를, value는 해당 소인수의 지수를 나타낸다.

    ex) `i = 18` -> `2^1 * 3^2` -> `{2:1, 3:2}`


```python
from collections.abc import Generator

class Factorization(Generator):
    pass

fact = Factorization(20)
for elem in fact:
    print(elem)
```

````python
from collections.abc import Generator
from collections import Counter
 
class Factorization():
    def __init__(self, n):
        self.n = n # 클래스 생성 시 정수 n을 입력받음

    def __iter__(self): # 제너레이터
        for a in range(1, self.n+1): # 1~ n까지 실행
            yield self.number_of_divisors(a)

        
    def is_prime(self, x):
        factorized = list()

        for idx in range(2, x):
            if x % idx == 0:
                factorized.append(idx)
        return len(factorized) == 0 # 소수 판별식


    def number_of_divisors(self, x):
      if x == 1:
        return None

      else:
        factorized_result = []
        answer = []

        for num in range(2, x+1):
          if self.is_prime(num) == True:
            factorized_result.append(num) # x 까지의 값 중 소수만 List 담기
        
        
        for numbers in factorized_result:
            # 1) 소인수 찾기
            if x % numbers == 0:
              while x % numbers == 0:
                answer.append(numbers)
                x = x // numbers

        # 2) dictionary로 소인수 & 지수 출력
        return dict(Counter(answer))           


 
fact = Factorization(20)
for elem in fact:
    print(elem)
````

````python
 # 과제 4번 코드란

def solution(a):
    answer_list = []
    for first in range(len(a)):
      for second in range(first+1, len(a)):
        for third in range(second+1, len(a)):
          #print(a[first] + a[second] + a[third])
          if (a[first] + a[second] + a[third]) == 0:
            answer_list.append([a[first], a[second], a[third]])
    return answer_list
 
print(solution([4, -2, 8, 4, -6, 12]))
print(solution([-25, -10, -7, -3, 2, 4, 8, 10]))

````

- 수업시간에 배운 제너레이터를 클래스로 구현하는 문제였습니다.
- 아래 답안 예제코드와 테스트용 코드를 첨부하니, 참고 부탁드립니다 :)

```python
from collections.abc import Generator

class Factorization(Generator):
    def __init__(self, n):
        self.n = n
        self.i = 0
        self.primes = self.find_prime(n)

    @staticmethod
    def find_prime(num):
        primes = list()
        for n in range(2, num + 1):
            test = 0
            for i in range(2, n):
                if n % i == 0:
                    test += 1
                    break
            if test == 0:
                primes.append(n)
        return primes

    def factorize(self, num):
        if num == 1:
            return dict()
        
        sol = dict()
        for prime in self.primes:
            if prime > num:
                break
        
            if num % prime == 0:
                temp = num
                exp = 0
                while temp % prime == 0:
                    exp += 1
                    temp /= prime
                sol[prime] = exp
        
        return sol

    def send(self, value):
        if self.i < self.n:
            result = self.i
            self.i += 1
            return self.factorize(self.i)
        else:
            self.throw(StopIteration)

    def throw(self, type, value=None, traceback=None):
        raise type

fact = Factorization(20)
for elem in fact:
    print(elem)
```
